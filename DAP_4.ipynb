## Q-4

```python
data_matrix = np.array([ [ 4,  7],
       [ 6,  9],
       [ 8, 10],
       [ 5,  8],
       [ 7, 11]])
data_matrix
```

```python
transpose_matrix = data_matrix.T
transpose_matrix
```

```python
# addition = data_matrix + transpose_matrix
# Matrix addition of a matrix and its transpose is defined only if the matrix is square.
```

```python
product_matrix = np.dot(data_matrix,transpose_matrix)
product_matrix
```

```python
#'''A matrix is invertible if: It is square & Determinant â‰  0'''
```

```python
# Only square matrices can have inverses
inv_matrix = np.linalg.inv(product_matrix)
inv_matrix
```

```python
row_sum = data_matrix.sum(axis=1)
column_sum = data_matrix.sum(axis=0)
print('Row sums of data_matrix',row_sum)
print('Column sums of data_matrix',column_sum)
```

```python
eigen_values , eigen_vectors = np.linalg.eig(product_matrix)
eigen_values
eigen_vectors
```
