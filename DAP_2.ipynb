# Q-1

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

product_A = [120, 135, 150, 160, 170, 180, 175, 165, 155, 145, 140, 130]
product_B = [100, 110, 115, 130, 140, 150, 160, 155, 145, 135, 125, 115]
product_C = [90, 95, 100, 105, 110, 120, 125, 130, 128, 120, 110, 100]
product_D = [80, 85, 90, 95, 100, 105, 110, 108, 102, 95, 90, 85]

plt.figure()
plt.bar(months,product_A)
plt.xlabel('Month')
plt.ylabel('Sales unit')
plt.title('monthly sales of Product A')
plt.savefig("2.1.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Q-2

```python
plt.figure()
plt.hist(product_B,bins=5,edgecolor ='black') # How to find number of bins : 1 = log 2n 
plt.xlabel('Sales unit')
plt.ylabel('Frequency')
plt.title('sales distribution of Product B')
plt.savefig("2.2.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Q-3

```python
plt.figure()
totals = (sum(product_A) , sum(product_B) ,sum(product_C), sum(product_D))
labels=('Product A','Product B','Product C','Product D')
plt.pie(totals,labels=labels,autopct='%1.1f%%')
plt.title('Annual sales contribution')
plt.savefig("2.3.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Q-4 

```python
plt.figure()
plt.plot(months,product_C,marker='o')
plt.xlabel('Month')
plt.ylabel('Sales units')
plt.title('Monthly sales trend of product C')
plt.savefig("2.4.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Q-5

```python
plt.figure()
plt.plot(months,product_A,label='Product A',marker='o')
plt.plot(months,product_B,label='Product B',marker='o')
plt.plot(months,product_C,label='Product C',marker='o')
plt.plot(months,product_D,label='Product D',marker='o')
plt.xlabel('Month')
plt.ylabel('Sales units')
plt.title('Monthly line chart of Products A,B,C,D')
plt.savefig("2.5.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Q-6

```python
plt.figure()
plt.bar(months,product_A,label='Product A')
plt.bar(months,product_B,bottom=product_A,label='Product B')
plt.xlabel('Month')
plt.ylabel('sales units')
plt.title('Stacked bar chart of Product A and B')
plt.legend()
plt.savefig("2.6.png", dpi=300, bbox_inches="tight")
plt.show
```

# Q-7

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.bar(months,product_A)
plt.title('Product A sales')

plt.subplot(2,2,2)
plt.hist(product_A,bins=5,edgecolor='black')
plt.title('Product B distribution')

plt.subplot(2,2,3)
plt.plot(months,product_C,marker='o')
plt.title('Product C sales trend')

plt.subplot(2,2,4)
plt.bar(months,product_A,label='Product A')
plt.bar(months,product_B,bottom=product_A,label='Product B')
plt.title('Stacked sales of A and B')
plt.legend()

plt.tight_layout()
plt.savefig("my_graph.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Q-8

```python
def f(x):
    return x**2

x = range(-10,11)
y = [f(i) for i in x]

plt.figure()
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph for function f(x) = x^2')
plt.savefig("2.8.png", dpi=300, bbox_inches="tight")
plt.show()
```
