## Q-1

```python
def clean_data(data):
    y = []
    for i in data:
        if i != None : y.append(i) 
        
    return y

cleaned_data = clean_data(study_hours)
print('Cleaned data :',cleaned_data) 
print('length of cleaned data is :',len(cleaned_data))
```
