# Q-4

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]

m=[]
for i in range(0,5):
    m.append(x[i] + y[i])
print('Element wise addition ',m)

m_zip=[]
for a,b in zip(x,y):
    m_zip.append(a+b)
print('Addition using zip ',m_zip)


    
    
```

# Q-6

```python
data=[10,None,20,30,None,40]

print('Number of missing values',data.count(None))

cleaned_data=[]
for i in data:
    if i!= None : cleaned_data.append(i)
    else :continue
print('Sum of cleaned data is',sum(cleaned_data))

```
